```python
import numpy as np
import os
import scipy.io as sio

if 1:
    home1= r"../data/源域数据集"
else:
    home1= r"../data/32kHz"
for root, dirs, files in os.walk(home1):
    for file in files:
        if file.endswith('.mat'):
            file_path = os.path.join(root, file)
            try:
                data = sio.loadmat(file_path)
                print(f"{file_path}\n\t{data.keys()}")
            except Exception as e:
                print(f"Could not read file {file_path}: {e}")
```
